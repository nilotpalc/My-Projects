```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_excel("Separate_Charts_Data.xlsx",sheet_name = "Sheet1" )
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["PLT_Lead_Time_Norm"] = dataset["PLT_Lead_Time_Norm"].map({"PLT Lead Time of 10":"# 4 PLTs in Month","PLT Lead Time of 7":"# 5 PLTs in Month",\
                                                                  "PLT Lead Time of 5":"# 7 PLTs in Month"})
```

```python
plt.rcParams.update({'font.size': 15})

g = sns.catplot(data=dataset,col="PLT_Lead_Time_Norm",row="Scenario",col_order=["# 5 PLTs in Month","# 7 PLTs in Month","# 4 PLTs in Month"],margin_titles=True,\
                   x="Plan",y="Std_Dev (PLT LT inc Delay)",order=["Plan A","Plan B"],ci=None,kind="bar",
                row_order=["Opg Stk of 50","Opg Stk of 100","Opg Stk of 150"])

g.set_titles(row_template = '{row_name}', col_template = '{col_name}')
```
