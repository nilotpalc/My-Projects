```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel("TRECS India plant reports JAN 2018 TO AUG 2020.xlsx",sheet_name="summary_report",header=0,usecols="A:Q,T:U")
```

```python
dataset[dataset['FolioID'] == 27210]
```

```python
dataset.drop_duplicates(subset=['FolioID','TankCode','TerminalProductCode','MonthNo','YearNo'],keep='last',inplace=True,ignore_index=True)
```

```python
dataset.sort_values(by="FolioID",ascending=True,ignore_index=True,inplace=True)
```

```python
dataset['Key'] = dataset['TankCode'] + dataset['TerminalProductCode'].astype(str) + dataset['MonthNo'].astype(str) + dataset['YearNo'].astype(str)
```

```python
zz = set(dataset['Key'].to_list())
```

```python
zz
```

```python
df = pd.DataFrame(columns = ['FolioID', 'PlantID', 'TankCode', 'TerminalProductCode',
       'TerminalProductDesc', 'OpeningStock', 'Receipt', 'Disposal',
       'ClosingStock', 'MonthNo', 'YearNo', 'Key', 'Counter',
       'Median_Value', 'FN_Counter'])
```

```python
df
```

```python
for i in zz:
    ab = dataset[dataset['Key'] == i ].reset_index().copy()
    ab['Counter'] = ab.index+1
    ab['Median_Value']=ab['Counter'].median()
    ab['FN_Counter'] = np.where(ab['Counter']<=ab['Median_Value'],'FN1','FN2')
    ab.drop(columns = ['index', 'OpeningStockTemp','OpeningStockDensity','BookStock','ClosingStockTemp', 'ClosingStockDensity',\
                       'Gain','Throughput', 'Tolerance'], axis=1,inplace = True)
    df = df.append(ab)
```
