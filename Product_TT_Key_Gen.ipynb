```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel("G2G Sep'20.xlsx",sheet_name="Sep'20",usecols = "E:N,AB:AC",header=1)
```

```python
dataset.replace(r'^\s*$',np.nan, regex=True,inplace=True)
dataset.fillna(0,axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.set_axis(['Month','Plant','Date of Loading','Truck No','Hauler','Product','ULP','SVPM','HSD','SVPD','Depot In','Depot Out'],
                axis='columns',inplace=True)
```

```python
dataset
```

```python
dataset['MS'] = dataset['ULP'] + dataset['SVPM']
dataset['AGO'] = dataset['HSD'] + dataset['SVPD']
```

```python
dataset.drop(['Hauler','Product','ULP','SVPM','HSD','SVPD'],axis=1,inplace=True)
```

```python
for i in range(len(dataset)):
    if isinstance(dataset.loc[i,'Month'],datetime.date) == False:
        dataset.loc[i,'Month'] = math.nan
        
    if isinstance(dataset.loc[i,'Date of Loading'],datetime.date) == False:
        dataset.loc[i,'Month'] = math.nan
        
    if isinstance(dataset.loc[i,'Depot In'],datetime.time) == False:
        dataset.loc[i,'Month'] = math.nan
        
    if isinstance(dataset.loc[i,'Depot Out'],datetime.time) == False:
        dataset.loc[i,'Month'] = math.nan
```

```python
dataset.dropna(axis=0,inplace=True)
```

```python
dataset.columns
```

```python
zz = dataset.groupby(['Month', 'Plant', 'Date of Loading', 'Truck No', 'Depot In',
       'Depot Out']).agg({'MS':sum,'AGO':sum}).reset_index()
```

```python
zz['Prod Type'] = np.where((zz['MS']>0) & (zz['AGO']>0),"MPL","Single")
```
